```python
#### original code working #######################################################################
data_path = "Excel_data/Roughness_DENSE.csv" 
data = pd.read_csv(data_path)
data_array=data.to_numpy()
x=data_array[:,0:6].astype(np.float32)
y1=data_array[:,6].reshape(-1,1) ######## shape is (275,1)
encoder = OneHotEncoder(sparse=False)
y = encoder.fit_transform(y1)
#print('encoded',y)

# Split the data for training and testing
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.20,shuffle=True, random_state=0)
print(train_x.shape)
print(train_x[0].shape)
# Build the model
model = Sequential()
model.add(Dense(20, input_shape=train_x[0].shape, activation='relu', name='fc1'))
model.add(Dense(15, activation='relu', name='fc2'))
model.add(Dense(11, activation='softmax', name='output'))
# Adam optimizer with learning rate of 0.001
optimizer = Adam(lr=0.001)
model.compile(optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
#print('Neural Network Model Summary: ') 
#print(model.summary())

# Train the model
#history=model.fit(train_x, train_y, verbose=2, batch_size=1, epochs=200)### works well for low batch size
history=model.fit(train_x, train_y, validation_data=(test_x, test_y), verbose=2, batch_size=1, epochs=200)### works well for low batch size
# Test on unseen data

results = model.evaluate(test_x, test_y)

print('Final test set loss: {:4f}'.format(results[0]))
print('Final test set accuracy: {:4f}'.format(results[1]))


```

```python
### TESTING CODE ###########################################################################
data_path = "Excel_data/Roughness_DENSE.csv" 
data = pd.read_csv(data_path)
data_array=data.to_numpy()
x=data_array[:,0:6].astype(np.float32)
y1=data_array[:,6].reshape(-1,1) ######## shape is (275,1)
encoder = OneHotEncoder(sparse=False)
y = encoder.fit_transform(y1)
print('encoded',y)

# Split the data for training and testing
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.20,shuffle=True, random_state=5)

# Build the model
model = Sequential()
model.add(Dense(64, input_shape=train_x[0].shape, activation='relu', name='fc1'))
model.add(BatchNormalization()) ## adding these are reducing scores for batch size 1 but increases for the acc for batch size 32
model.add(Dense(32, activation='relu', name='fc2'))
model.add(BatchNormalization()) 
model.add(Dense(11, activation='softmax', name='output'))
model.summary()
# Adam optimizer with learning rate of 0.001
optimizer = Adam(lr=0.001)
model.compile(optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
#model.compile(optimizer, loss=tf.keras.losses.BinaryCrossentropy(from_logits=True), metrics=['accuracy'])

# Train the model
history=model.fit(train_x, train_y, verbose=2, batch_size=32, epochs=100)
# Test on unseen data
results = model.evaluate(test_x, test_y)
print('Final test set loss: {:4f}'.format(results[0]))
print('Final test set accuracy: {:4f}'.format(results[1]))
model.save('dense_rough_same_param_model_25.h5')
```

```python
##### ACCURACY ######################
ax = plt.axes()
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(len(acc))
plt.plot(epochs, acc, 'r', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Test acc')
plt.title('Training and Test Accuracy',fontsize=20)
plt.ylabel('Accuracy',fontsize=20) 
plt.xlabel('Epoch',fontsize=20)
plt.legend()
plt.tick_params(axis='both', labelsize=12)
plt.savefig("rough_only_ACCgraph.pdf",bbox_inches='tight')
plt.savefig("rough_only_ACCgraph.eps",bbox_inches='tight')

###### LOSS ################################
plt.figure()
plt.plot(epochs, loss, 'r', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Test loss')
plt.title('Training and Test Loss',fontsize=20)
plt.ylabel('Loss',fontsize=20) 
plt.xlabel('Epoch',fontsize=20)
plt.legend()
plt.tick_params(axis='both', labelsize=12)
plt.savefig("rough_only_loss.pdf",bbox_inches='tight')
plt.savefig("rough_only_loss.eps",bbox_inches='tight')
```
